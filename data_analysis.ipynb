```python
from tqdm import tqdm
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from dataset import UNSW_NB15
```

```python
legend = "data/NUSW-NB15_features.xlsx"
files = ["data/UNSW_NB_15_1_withCVSS_V2.csv", 
         "data/UNSW_NB_15_2_withCVSS_V2.csv",
         "data/UNSW_NB_15_4_withCVSS_V2.csv"]

feat_legend = pd.read_excel(legend, header=0, engine='openpyxl')
dtypes = feat_legend['Type '].str.lower()
names = feat_legend['Name'].str.lower()

frame = pd.DataFrame()
for file in files:
    df = pd.read_csv(file, header=0, dtype=object, names=names)
    frame = pd.concat([frame, df], axis=0)

numericals = np.where(dtypes != 'nominal')[0]
for idx in tqdm(numericals, desc="Curing Data"):
    frame.iloc[:, idx] = pd.to_numeric(frame.iloc[:, idx], errors='coerce') 
```

```python
column = frame.iloc[:, 3]

ax = column.dropna().hist(bins=200)
ax.cumulative=True
ax.set_xlim([0, max(column)])
ax.set_ylim([0, 0.2e5])
plt.title(f'Histogram of {column.name}.')
```

```python
column = frame.iloc[:, 3]

ax = column.plot.density()
ax.set_xlim([0, max(column)])
plt.title(f'Probability Distribution of {column.name}.')
```

```python
for i, ft in enumerate(frame.columns):
    num_opt = len(frame[ft].dropna().unique())
    num_missing = sum(frame[ft].isna())
    num_minority = frame[ft].value_counts(dropna=True, ascending=True).iloc[0]
    type = dtypes.iloc[i]

    if type != 'nominal':
        ax = frame[ft].plot.density()
        ax.set_xlim([0, max(frame[ft])])
        plt.title(f'Probability Distribution of {ft}.')    
        plt.show()

    print(f'Name: {ft}')
    print(f'Cardinality: {num_opt}')
    print(f'Percentage Constituted by Minority Class: {100* num_minority / len(frame)}')
    print(f'Percentage Missing: {100 * num_missing / len(frame)}')
    print(f'Type: {type}\n')
```

```python
max_bins = 200
stats_table = {}

for i, ft in enumerate(frame.columns):
    num_opt = len(frame[ft].dropna().unique())
    num_missing = sum(frame[ft].isna())
    num_minority = frame[ft].value_counts(dropna=True, ascending=True).iloc[0]
    type = dtypes.iloc[i]

    bins = num_opt if num_opt < max_bins else max_bins
    frame[ft].dropna().hist(bins=bins)
    if type == 'nominal':
        plt.xticks([])
    plt.title(f'{ft}')
    plt.xlabel('Value')
    plt.ylabel('Number of Occurrences')
    plt.show()

    print(f'Name: {ft}')
    print(f'Cardinality: {num_opt}')
    print(f'Percentage Constituted by Minority Class: {100* num_minority / len(frame)}')
    print(f'Percentage Missing: {100 * num_missing / len(frame)}')
    print(f'Type: {type}\n')
    
    stats_table[ft] = [num_opt, num_minority / len(frame), num_missing / len(frame), type]

stats_table = pd.DataFrame(stats_table, index=['Cardinality', 'Percentage Constituted by Minority Class', 'Percentage Missing', 'Type'])
```
